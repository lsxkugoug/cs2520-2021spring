```python
# This script is used to setUp, update, and deploy spines overlay network
# if want to see the logs of receiver spine,the script will dismiss the receiver node. and you need use deployReceiver_log notebook
# to deploy receiver' nodes spines
```

```python
from os import kill
import paramiko
```

```python
# flag parameters
nodeFile = 'node.txt'
sshUserName = 'shl199'
pwd = 'Lsx314159'
setUpFlag = True    # if the machine is origin, clone the repo and execute ./configure
hFileChanged = False # if .h file changed, should make clean in advance
keyFilePath = './id_geni_ssh_rsa'
receiverFlag = True # if want to use deployReceiver_log to see the receiver's spines log
reciverNode = 'pcvm5-3.instageni.idre.ucla.edu'
```

```python
# global variable definition
sshCon = [] # store the connections
```

```python
# read the nodes
nodes = None
with open(nodeFile, "r") as fd:
    nodes = fd.read().splitlines()
print("Node list readed as:")
print(nodes)
```

```python
# install and compile code include reciver node regardless about receiverFlag，if not need that, skip it and execute deploy spines part
for nodePort in nodes:
    tmp = nodePort.split()
    node = tmp[0]
    port = 22
    if len(tmp) == 2:
        port = tmp[1]
    print("process node:{}, port:{}".format(node,port))
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    ssh.connect(node, username=sshUserName, password=pwd ,key_filename= keyFilePath, port=port)
    sshCon.append(ssh)
    if setUpFlag == True:
        print("execute set up process, git clone and ./configure")
        ssh.exec_command("git clone https://github.com/lsxkugoug/cs2520-2021spring.git;")
        ssh.exec_command('cd ~/cs2520-2021spring/p4/spines-src-5.5; ./configure;')
    if hFileChanged == True:
        print("./h file changed, make clean")
        ssh.exec_command('cd ~/cs2520-2021spring/p4/spines-src-5.5; make clean;')
    print('compile code')
    stdin, stdout, stderr = ssh.exec_command('cd ~/cs2520-2021spring/p4/spines-src-5.5;git pull ; make -C daemon parser; make;')
    print(stdout.readlines())

for ssh in sshCon:
    ssh.close()
sshCon.clear()
```

```python
# deploy spines, nohup can allow process in background 
for nodePort in nodes:
    tmp = nodePort.split()
    node = tmp[0]
    port = 22
    if len(tmp) == 2:
        port = tmp[1]
    if receiverFlag == True and node == reciverNode:
        print('dismiss receiver spines, please use deploy_Spines_execeptRec to deploy reciever node')
        continue
    print("process node:{}, port:{}".format(node,port))
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    ssh.connect(node, username=sshUserName, password=pwd ,key_filename= keyFilePath, port=port)
    ssh.exec_command('cd ~/cs2520-2021spring/p4/spines-src-5.5/daemon;git pull; nohup ./spines  -m -w problem')
    sshCon.append(ssh)
```

```python
# kill the spines and release connection
for nodePort in nodes:
    tmp = nodePort.split()
    node = tmp[0]
    port = 22
    if len(tmp) == 2:
        port = tmp[1]
    if receiverFlag == True and node == reciverNode:
    print('dismiss receiver spines, please use deploy_Spines_execeptRec to deploy reciever node')
    continue
    print("kill node:{}, port:{}".format(node,port))
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    ssh.connect(node, username=sshUserName, password=pwd ,key_filename= keyFilePath, port=port)
    ssh.exec_command('killall -u {}'.format(sshUserName))
```
